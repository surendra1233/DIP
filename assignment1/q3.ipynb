```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
```

# Part-1

```python
def BitQuantizeImage(im,k):
    b = 2**k -1
    out = im*(b/255)
    return out.astype(np.uint8)
```

```python
img = cv2.imread('./DIP_2019_A1/quantize.jpg')
out = BitQuantizeImage(img,6)
out_rgb = cv2.cvtColor(out,cv2.COLOR_BGR2RGB)
img_rgb = cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
fig = plt.figure()
fig.add_subplot(121)
plt.title('original')
plt.imshow(img_rgb) 
plt.subplot(1,2,2)
plt.title('after Bit Quantization')
plt.imshow(out_rgb)
```

If we see here, in the histograms
+ For **lena1.png**, we have all values at 0 and 255. It should be MSB bit-plane or binary image scaled to 0-255.If we check with the images it is **MSB bitplane**.
+ For **lena2.png**, we have all values at 0,64,128,192. So we can say that this is the **combination of 7th and 6th bitplanes** (2^7+2^6)
+ For **lena3.png**, we have all values at 0 and 255. It should be MSB bit-plane or binary image scaled to 0-255.If we check with the images it is **binary image**.
