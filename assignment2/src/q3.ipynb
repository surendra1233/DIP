```python
import numpy as np
import matplotlib.pyplot as plt
import cv2
```

## Part-1

```python

def convolution2d(img, fil, s=1, padding=True):
    img = img.astype(np.float64)
    fil_size = fil.shape[0]
    if padding:
        pad_size = fil_size//2
        img = np.pad(img,((pad_size,pad_size)),mode='constant')

    h,w = img.shape[0],img.shape[1]
    
    rows = (h - fil_size)//s +1
    cols = (w - fil_size)//s +1
    out = np.zeros((rows,cols))
    
    
    for i in range(rows):
        for j in range(cols):
            curr = img[i:i+fil_size,j:j+fil_size]
            out[i,j] = np.sum(np.multiply(curr,fil))
    
    out = np.clip(out,0,255)
    return out.astype(np.uint8) 
```

```python
img = cv2.cvtColor(cv2.imread('../A2_resources/box.png'),cv2.COLOR_BGR2GRAY)

filt = np.array([[0,-1,0],[0,0,0],[0,1,0]])
out = convolution2d(img,filt)
plt.imshow(out,'gray')
plt.show()
# print(np.argmax(img)/img.shape[0])
```

## Part-2

```python
img = cv2.cvtColor(cv2.imread('../A2_resources/blur.jpg'),cv2.COLOR_BGR2GRAY)
filt = np.array([[-1,-1,-1],[0,0,0],[1,1,1]])
out1 = convolution2d(img,filt)
out2 = convolution2d(img,filt.T)
fig = plt.figure()
fig.set_size_inches(20, 10)
fig.add_subplot(131)
plt.imshow(img,'gray')
plt.subplot(1,3,2)
plt.imshow(out1,'gray')
plt.subplot(1,3,3)
plt.imshow(out2,'gray')
plt.show()
```

We can see the **vertical edges** are getting detected when blur.jpg is convolved with the filter while **horizontal edges** are getting detected when blur.jpg is convolved with filter's transpose.
