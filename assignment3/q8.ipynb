```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
```

```python
img = cv2.cvtColor(cv2.imread('DIP_2019_A3/rectangle.jpg'),cv2.COLOR_BGR2GRAY)
a = np.concatenate((img[100:,:],img[:100,:]))

f1 = np.log(1+np.fft.fftshift(np.fft.fft2(img)))
f2 = np.log(1+np.fft.fftshift(np.fft.fft2(img.T)))
f3 = np.log(1+np.fft.fftshift(np.fft.fft2(a)))

plt.imshow(np.abs(f1),'gray')
plt.show()
plt.imshow(np.abs(f2),'gray')
plt.show()
plt.imshow(np.abs(f3),'gray')
plt.show()
```
