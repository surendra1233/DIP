```python
import numpy as np
import matplotlib.pyplot as plt
import cv2
```

```python
def nyquist(img,nx,ny):
    a,b=img.shape
    row = np.linspace(0,a-1,a//ny).astype(np.int64)
    col = np.linspace(0,b-1,b//nx).astype(np.int64)
    img=img[row,:]
    img=img[:,col]
    plt.imshow(cv2.resize(img,(a,b)),'gray')
```

```python
img = cv2.cvtColor(cv2.imread("./DIP_2019_A3/bricks.jpg"),cv2.COLOR_BGR2GRAY)
n_arr =[2,3,4,5,6]
for nx in n_arr:
    i = 0
    plt.figure(figsize=(20,20))
    for ny in n_arr:
        plt.subplot(len(n_arr),len(n_arr),i+1)
        nyquist(img,nx,ny)
        i+=1
plt.show()
```

width of distance b/w bricks in x direction merges after nx = 4 => nyquist sampling rate in x-direction = 333/4 = 83<br>
width of distance b/w bricks in y direction  is ny = 5 => nyquist sampling rate in y-direction = 413/5 = 83
nyquist sampling rate = min(83,83) = 83.

```python
img = cv2.filter2D(img,-1,np.ones([7,7])/49)
n_arr =[2,3,4,5,6]
for nx in n_arr:
    i = 0
    plt.figure(figsize=(20,20))
    for ny in n_arr:
        plt.subplot(len(n_arr),len(n_arr),i+1)
        nyquist(img,nx,ny)
        i+=1
plt.show()
```

Here we can see that nyquist rate decreased as the image is blurred. Because when nx = 6,ny=6 bricks are still seperable. So we can get a lesser minimum sampling rate.
