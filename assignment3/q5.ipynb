```python
import numpy as np
import cv2
from scipy import signal
import matplotlib.pyplot as plt
```

```python
img = cv2.cvtColor(cv2.imread('DIP_2019_A3/lena.jpg'),cv2.COLOR_BGR2GRAY)
filt = np.array([[0,1,0],[1,2,1],[0,1,0]])
a = np.zeros(img.shape)
a[:3,:3] = filt
fil_fft = np.fft.fft2(a)
out = abs(np.fft.ifft2(fil_fft*np.fft.fft2(img)))
plt.imshow(np.log(1+np.abs(np.fft.fftshift(fil_fft))),'gray')
plt.show()
plt.imshow(out,'gray')
plt.show()
```
