```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
```

```python
def fft_1d(x):
    x = np.asarray(x, dtype=float)
    N = x.shape[0]
    if N==1:
        return x

    X_even = fft_1d(x[::2])
    X_odd = fft_1d(x[1::2])
    factor = np.exp(-2j * np.pi * np.arange(N//2) / N)
    factor = factor.reshape((factor.shape[0],1))
    a = np.asarray(X_even + factor * X_odd)
    b = np.asarray(X_even - factor * X_odd)
    return np.concatenate((a,b))

def fft_2d(img):
    img = np.asarray(img, dtype=float)
    a = fft_1d(img)
    return fft_1d(a.T).T
```

```python
img = cv2.cvtColor(cv2.imread('DIP_2019_A3/lena.jpg'),cv2.COLOR_BGR2GRAY)
out = np.log(1+np.fft.fftshift(fft_2d(img)))
plt.imshow(np.abs(out),'gray')
plt.show()
```

```python
img = cv2.cvtColor(cv2.imread('DIP_2019_A3/lena.jpg'),cv2.COLOR_BGR2GRAY)
out = np.log(1+np.fft.fftshift(np.fft.fft2(img)))
plt.imshow(np.abs(out),'gray')
plt.show()
```
