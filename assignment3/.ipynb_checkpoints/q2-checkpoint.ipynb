# Part-1

```python
import numpy as np
import matplotlib.pyplot as plt
import cv2
import numpy.fft as fft
```

```python
def ideal_lowpass(img,rad=20):
    ff = fft.fft2(img)
    sff = fft.fftshift(ff)
    a,b = [sff.shape[0]//2,sff.shape[1]//2]
    i = np.arange(sff.shape[0])
    j = np.arange(sff.shape[1])
    
    i = np.matlib.repmat(i,sff.shape[1],1).T
    j = np.matlib.repmat(j,sff.shape[0],1)
    
    sff[(i-a)**2 + (j-b)**2>rad*rad] = 0
    out = fft.ifft2(fft.ifftshift(sff)).real

    plt.figure(figsize=(20,10))
    plt.subplot(1,3,1)
    plt.imshow(np.abs(np.log(1+sff)),'gray')
    plt.subplot(1,3,2)
    plt.imshow(img.astype(np.uint8),'gray')
    plt.subplot(1,3,3)
    plt.imshow(out.astype(np.uint8),'gray')
    return out

def butterworth_lowpass(img,d = 20,n=1,):
    ff = fft.fft2(img)
    sff = fft.fftshift(ff)
    a,b = [sff.shape[0]//2,sff.shape[1]//2]
    i = np.arange(sff.shape[0])
    j = np.arange(sff.shape[1])
    
    i = np.matlib.repmat(i,sff.shape[1],1).T
    j = np.matlib.repmat(j,sff.shape[0],1)
    
    filt = 1/(1+(((i-a)**2 + (j-b)**2)/(d*d))**n)
    sff = np.multiply(filt,sff)
    out = fft.ifft2(fft.ifftshift(sff)).real
    
    plt.figure(figsize=(20,10))
    plt.subplot(1,3,1)
    plt.imshow(np.abs(np.log(1+sff)),'gray')
    plt.subplot(1,3,2)
    plt.imshow(img.astype(np.uint8),'gray')
    plt.subplot(1,3,3)
    plt.imshow(out.astype(np.uint8),'gray')
    return out
def gaussian_lowpass(img,sigma):
    ff = fft.fft2(img)
    sff = fft.fftshift(ff)
    a,b = [sff.shape[0]//2,sff.shape[1]//2]
    i = np.arange(sff.shape[0])
    j = np.arange(sff.shape[1])
    i = np.matlib.repmat(i,sff.shape[1],1).T
    j = np.matlib.repmat(j,sff.shape[0],1)
    
    filt = np.exp(-((i-a)**2+(j-b)**2)/(2*sigma*sigma))
    
    sff = np.multiply(filt,sff)
    out = fft.ifft2(fft.ifftshift(sff)).real
    
    plt.figure(figsize=(20,10))
    plt.subplot(1,3,1)
    plt.imshow(np.abs(np.log(1+sff)),'gray')
    plt.subplot(1,3,2)
    plt.imshow(img.astype(np.uint8),'gray')
    plt.subplot(1,3,3)
    plt.imshow(out.astype(np.uint8),'gray')
    return out
```

```python
img = cv2.cvtColor(cv2.imread("./DIP_2019_A3/lena.jpg"),cv2.COLOR_BGR2GRAY)
out = ideal_lowpass(img,100)
out = butterworth_lowpass(img,10)
out = gaussian_lowpass(img,10)
```

# Part-2

```python
out1=gaussian_lowpass(img,40)
out2=gaussian_lowpass(img,10)
plt.figure()
plt.imshow(np.abs(out1-out2),'gray')
plt.show()
```

The difference between the two gaussians gives(approximates) the laplacian of gaussian of the image. Here we can see that we got high-frequency(edge information) output of the image which is the same as laplacian of gaussian of the image.
