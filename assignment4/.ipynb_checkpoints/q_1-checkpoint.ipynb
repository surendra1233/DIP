```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
def hitandmiss(img,s):
    final = np.zeros((img.shape[0]-2,img.shape[1]-2))
    for i in range(1,img.shape[0]-1):
        for j in range(1,img.shape[1]-1):
            this = 0
            for k in range(-1,2):
                for l in range(-1,2):
                    if this!=1 and s[k+1][l+1]!='*' and np.float64(s[k+1][l+1])!=img[i+k][j+l]:
                        this = 1
            if this==0:
                final[i-1][j-1] = this
            else:
                final[i-1][j-1] = img[i,j]
    plt.figure(figsize=(6,6))
    plt.imshow(final,cmap="gray")
    plt.show()
    return final
    
def skeletonization(img,s1,s2):
    before = np.ones(img.shape)
    final = img
    i=0
    s1 = np.rot90(s1,3)
    s2 = np.rot90(s2,3)
    while not np.array_equal(before,final):
        print(i)
        before = np.copy(final)
        nimg = np.zeros((before.shape[0]+2,before.shape[1]+2))
        nimg[1:before.shape[0]+1,1:before.shape[1]+1] = before
        if i%2==0:
            s1 = np.rot90(s1,1)
            final = hitandmiss(nimg,s1)
        else:
            s2 = np.rot90(s2,1)
            final = hitandmiss(nimg,s2)
        i = i + 1
        if(i==8): 
            i=0
        
    return final
            
bio = cv2.imread("../input_data/biometric.jpg",0)
plt.figure(figsize=(6,6))
plt.imshow(bio,cmap="gray")
plt.show()
ret,thresh = cv2.threshold(bio,127,255,cv2.THRESH_BINARY)
thresh = thresh/255
newthresh = np.copy(thresh)
thresh[newthresh==0] = 1
thresh[newthresh==1] = 0
s1 = np.array([[0,0,0],['*',1,'*'],[1,1,1]])
s2 = np.array([['*',0,0],[1,1,0],['*',1,'*']])
plt.figure(figsize=(6,6))
plt.imshow(thresh,cmap="gray")
plt.show()
# print(((thresh==1).sum())+(thresh==0).sum())
# print(thresh.shape[0]*thresh.shape[1])
finalimg = skeletonization(thresh,s1,s2)
plt.figure(figsize=(6,6))
plt.imshow(finalimg,cmap="gray")
plt.show()
```
